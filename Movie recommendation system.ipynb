```python
import numpy as np
import pandas as pd
import sklearn
import os
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
l1 = [1,1,0,1,1]
l2 = [0,1,0,1,1]
print(cosine_similarity([l1, l2]))
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
os.chdir('C:\\Users\\roger\\Downloads\\Kaif\\')
movies =pd.read_csv('movies data.csv')
movies.head()
main_features = ['genres','keywords','tagline','cast','director']
print(main_features)
print(movies.info())
print(movies.isna().sum())
print(movies[main_features].head())
print(movies[main_features].isna().sum())
for feature in main_features:
  movies[feature] = movies[feature].fillna('')
print(movies.head())
mixed_features = movies['genres']+' '+movies['keywords']+' '+movies['tagline']+' '+movies['cast']+' '+movies['director']
print(mixed_features)
vect = TfidfVectorizer()
feature_vectors = vect.fit_transform(mixed_features)
print(feature_vectors.shape)
print (feature_vectors)
common = cosine_similarity(feature_vectors)
print  (common)
print(common.shape)
movie_name = input(' Enter your favourite movie name : ')
list_titles = movies['title'].tolist()
print(list_titles)
close_match = difflib.get_close_matches(movie_name, list_titles)
print(close_match)
close_match = close_match[0]
print(close_match)
index_movie = movies[movies.title == close_match]['index'].values[0]
print(index_movie)
similarity_score = list(enumerate(common[index_movie]))
print(similarity_score)
len(similarity_score)
similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(similar_movies)
print('Movies suggested for you : \n')
i = 1
for movie in similar_movies:
  index = movie[0]
  title = movies[movies.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title)
    i+=1
```
